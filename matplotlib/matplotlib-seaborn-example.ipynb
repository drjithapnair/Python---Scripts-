```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Load dataset
iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
# Create scatter plot
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=iris)
# Add labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Scatter Plot of Sepal Length vs Petal Length")
# Show plot
plt.show()
```

```python
df1=iris.drop("species", axis="columns")
# Compute correlation matrix
corr = df1.corr()
corr
```

```python
# Create heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
# Add title
plt.title("Correlation Matrix of Iris Dataset")
# Show plot
plt.show()
```

```python
# Create violin plot
sns.violinplot(x="species", y="petal_length", data=iris)
# Add title
plt.title("Distribution of Petal Length by Species")
plt.show()
```

```python
# Create pair plot
sns.pairplot(iris, hue="species", diag_kind="hist")
# Show plot
plt.show()
```

```python
# Generate some sample data

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Matplotlib example
plt.figure(figsize=(10, 4))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Matplotlib: Sin and Cos Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Seaborn example
sns.set_style("darkgrid")
plt.figure(figsize=(10, 4))
sns.lineplot(x=x, y=y1, label='sin(x)')
sns.lineplot(x=x, y=y2, label='cos(x)')
plt.title('Seaborn: Sin and Cos Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Seaborn statistical plot example
tips = sns.load_dataset("tips")
plt.figure(figsize=(8, 5))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Seaborn: Box Plot of Total Bill by Day')
plt.show()
```

```python
# Create violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
# Add title
plt.title("Seaborn: violin Plot of Total Bill by Day")
plt.show()
```

```python
# Create pair plot
sns.pairplot(tips, hue="day", diag_kind="hist")
# Show plot
plt.show()
```
