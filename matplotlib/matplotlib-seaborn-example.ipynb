```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Matplotlib example
plt.figure(figsize=(10, 4))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Matplotlib: Sin and Cos Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Seaborn example
sns.set_style("whitegrid")
plt.figure(figsize=(10, 4))
sns.lineplot(x=x, y=y1, label='sin(x)')
sns.lineplot(x=x, y=y2, label='cos(x)')
plt.title('Seaborn: Sin and Cos Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Seaborn statistical plot example
tips = sns.load_dataset("tips")
plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Seaborn: Box Plot of Total Bill by Day')
plt.show()
```
