```python
# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
```

```python
# 1. Scatter plot with colormap (Matplotlib)
n = 1000
x = np.random.randn(n)
y = np.random.randn(n)
colors = np.random.rand(n)
sizes = 1000 * np.random.rand(n)
```

```python
scatter = axes[0, 0].scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
axes[0, 0].set_title('Matplotlib: Scatter Plot with Colormap')
fig.colorbar(scatter, ax=axes[0, 0], label='Color Value')
```

```python
# 2. Heatmap (Seaborn)
data = np.random.rand(10, 10)
sns.heatmap(data, annot=True, cmap='coolwarm', ax=axes[0, 1])
axes[0, 1].set_title('Seaborn: Heatmap')
```

```python
# 3. Multiple distributions (Seaborn)
iris = sns.load_dataset('iris')
sns.violinplot(x='species', y='sepal_length', data=iris, ax=axes[1, 0])
axes[1, 0].set_title('Seaborn: Violin Plot')
```

```python
# 4. Time series with error bands (Matplotlib)
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.2, 100)
axes[1, 1].plot(x, y, label='noisy sin(x)')
axes[1, 1].fill_between(x, y-0.2, y+0.2, alpha=0.3)
axes[1, 1].set_title('Matplotlib: Time Series with Error Bands')
axes[1, 1].legend()
```

```python
# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
# Bonus: Pair plot (Seaborn)
plt.figure(figsize=(12, 10))
sns.pairplot(iris, hue='species', height=2.5)
plt.suptitle('Seaborn: Pair Plot', y=1.02)
plt.show()
```
